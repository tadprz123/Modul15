```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

prices = [
    (1, 2.12),
    (2, 2.56),
    (3, 3.10),
    (4, 3.16),
    (5, 3.58),
    (6, 5.12),
    (7, 5.16),
    (8, 5.20),
    (9, 4.12),
    (10, 4.10),
    (11, 3.65),
    (12, 4.25),
]

df = pd.DataFrame(prices, columns=['Month', 'PricePLN'])
df.set_index('Month', inplace=True)
exchange_rate = 4
df['PriceUSD'] = df['PricePLN'] / exchange_rate
print(df)

plt.figure(figsize=(10, 6))
plt.plot(df.index, df['PriceUSD'], 'r--', marker='o') 
plt.xlabel('Miesiąc')
plt.ylabel('Cena w USD')
plt.title('Cena produktu w USD na przestrzeni miesięcy')
plt.grid(True)
plt.xticks(np.arange(1, 13)) 
plt.show()
```
